```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = np.array([12, 15, 20, 23, 25, 30, 32, 35, 38, 40, 45, 48, 50, 52, 55, 58, 60, 63, 65, 68, 70, 72, 75, 78, 80, 83, 85, 88, 90, 92])
data = np.sort(data)

bin_size = 5
num_bins = int(len(data) / bin_size)

bin_mean = np.zeros((num_bins, bin_size))
bin_median = np.zeros((num_bins, bin_size))
bin_boundary = np.zeros((num_bins, bin_size))

for i in range(0, len(data), bin_size):
    k = int(i / bin_size)
    mean_val = np.mean(data[i:i+bin_size])
    for j in range(bin_size):
        bin_mean[k, j] = mean_val

for i in range(0, len(data), bin_size):
    k = int(i / bin_size)
    median_val = np.median(data[i:i+bin_size])
    for j in range(bin_size):
        bin_median[k, j] = median_val


print("\n* Smoothing by Bin Mean:")
for i in range(num_bins):
    print(f"Bin{i+1}: {[str(int(x)) if x.is_integer() else str(x) for x in bin_mean[i]]}")

print("\n* Smoothing by Bin Median:")
for i in range(num_bins):
    print(f"Bin{i+1}: {[str(int(x)) if x.is_integer() else str(x) for x in bin_median[i]]}")
```
